```python
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
from telethon import TelegramClient, events, sync

# These example values won't work. You must get your own api_id and
# api_hash from https://my.telegram.org, under API Development.
api_id = os.getenv('mtprotoapiid')
api_hash = os.getenv('mtprotoapihash')

client = TelegramClient('user_session', api_id, api_hash)
```

```python
async with client:
    print(await client.get_me())
```

```python
posts = []
import datetime
import time

async with client:
    async for message in client.iter_messages("sb8blog"):
        # print(message.id, message.text)
        # print(message)
        dt = message.date.timestamp() * 1000
        date = message.date.strftime('%d/%m/%Y')
        title = ""
        text = message.text
        if not message.text:
            continue
        rows = message.text.split("\n")
        title = rows[0].replace("*", '')
        print(title)
        haslink = "https://" in text
        hasinstantview = "telegra.ph" in text
        posts.append({'id':message.id, 'title': title, 'text':message.text, 'date': date,
                      'haslink': haslink, 'hasinstantview': hasinstantview, 'MIS_DT': dt})
```

```python
import csv
import dateutil.parser as dparser

unsubs = []
with open("TGUnsubscriptions.csv", "r", encoding="utf-8", newline='') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in csvreader:
        if row[5] == "UnsubscriptionDate":
            continue
        # date = dparser.parse(row[4],fuzzy=True)
        unsubs.append(int(row[4]))
```

```python
import csv

with open('posts.csv', 'w', newline='', encoding="utf-8") as csvfile:
    fieldnames = ['id', 'title', 'text', 'date', 'MIS_DT', 'haslink', 'hasinstantview']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(posts)
```

```python
posts.sort(key=lambda x: x['id'])
for i in range(len(posts)):
    p = posts[i]
    p['periodstart'] = p['MIS_DT']
    p['periodend'] = datetime.datetime.now().timestamp() * 1000

    for j in range(i+1, len(posts)):
        p2 = posts[j]
        if (p2['MIS_DT'] == p['MIS_DT']):
            continue

        p['periodend'] = p2['MIS_DT']
        break
    print(i, p['periodstart'], p['periodend'])
```

```python
for p in posts:
    p['unsubs'] = 0
    for unsub in unsubs:
        if unsub >= p['periodstart'] and unsub < p['periodend']:
            p['unsubs'] += 1
    print(p['unsubs'])
```

```python
import csv

with open('posts.csv', 'w', newline='', encoding="utf-8") as csvfile:
    fieldnames = ['id', 'title', 'text', 'date', 'MIS_DT', 'haslink', 'hasinstantview', 'unsubs',
                  'periodstart','periodend']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(posts)
```
