Transformation for Online Retail Dataset I got on Kaggle: https://www.kaggle.com/datasets/carrie1/ecommerce-data

```python
import pandas as pd
import plotly.express as px
from random import random, choice

dsells = pd.read_csv("./data.csv")

def source(x):
    return choice(["Amazon", "Emag", "Website", "Offline"])

dsells["Source"] = dsells.index.map(source)

dsells.head()
```

```python
import numpy as np

dproducts = dsells.groupby(["Description", "Source"])\
    .agg({"Quantity": "sum", "InvoiceNo": "count", "UnitPrice": "mean"})\
    .rename(columns={"InvoiceNo": "Invoice Count"})

def randTop(x):
    return (random() > 0.2)

dproducts.reset_index(inplace=True)

dproducts["UnitPrice"] = dproducts["UnitPrice"].apply(lambda x: np.around(x, decimals=2) )
dproducts["TopPrice"] = dproducts.index.map(randTop)
dproducts.loc[dproducts["Source"] == "Offline", "TopPrice"] = True
dproducts.loc[dproducts["Source"] == "Website", "TopPrice"] = True

dproducts.head()
```

```python
dproducts.to_csv("transformed.csv")
```
