Transformation for generated Dataset in country_level.csv

```python
import pandas as pd
import plotly.express as px
from random import random, choice

dsells = pd.read_csv("./country_level.csv", sep=";")
dsells["Revenue"] = dsells["Revenue"].str.replace(',', '.').astype(float)
dsells["Profit"] = dsells["Profit"].str.replace(',', '.').astype(float)

dsells = dsells.rename(columns={"Orders cty": "Orders"})

dsells.head()
```

```python
import numpy as np

def randomize(x):
    return round(x * random())

def halved(x):
    return round(x * (0.5 + random() / 4))

dres = None

for y in range(2021, 2022+1):
    for m in range(1,12+1):
        dtemp = dsells.copy()
        dtemp["Month"] = m
        dtemp["Revenue"] = dtemp["Revenue"].apply(randomize)
        dtemp["Profit"] = dtemp["Revenue"].apply(randomize)
        dtemp["Profit"] = dtemp["Profit"].apply(halved)

        dtemp["Orders"] = dtemp["Orders"].apply(randomize)

        dtemp["Year"] = y

        if dres is None:
            dres = dtemp
        else:
            dres = pd.concat([dres, dtemp])

        
dres.sample(10)
```

```python
dres.to_csv("country_level_transformed.csv")
```
